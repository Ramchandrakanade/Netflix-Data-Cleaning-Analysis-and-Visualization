```python
#Step 1: Import Required Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# step 2 - creat the data frame 
# Load the dataset

dataframe = pd.read_csv('netflix_titles.csv')
print(dataframe)
```

```python
# Display the first few rows of the dataset

print(dataframe.head())
```

```python
# Step 3: Data Cleaning
# Identify and handle missing data, correct data types, and drop duplicates.


# Check for missing values
print(dataframe.isnull().sum())
```

```python
# Drop duplicates if any
dataframe.drop_duplicates(inplace=True)
print(dataframe)
```

```python
# Drop rows with missing critical information

# Since the 'cast' column does not exist in the dataframe, drop rows only based on 'director' and 'country'

dataframe.dropna(subset=['director', 'country'], inplace=True)
print(dataframe)
```

```python
# Convert 'date_added' to datetime format

dataframe['date_added'] = pd.to_datetime(dataframe['date_added'])
print(dataframe)
```

```python
# Show data types to confirm changes

print(dataframe.dtypes)
```

```python
#Step 4: Exploratory Data Analysis (EDA)
# This step would typically include visualizations and statistical summaries to understand the data better.

# Example: Display the distribution of content types

# 1. Content Type Distribution (Movies vs. TV Shows)
# Count the number of Movies and TV Shows

type_counts = dataframe['type'].value_counts()

# Plot the distribution
plt.figure(figsize=(8, 6))
sns.barplot(x=type_counts.index, y=type_counts.values,palette='Set2')
plt.title('Distribution of Content by Type')
plt.xlabel('Type')
plt.ylabel('Count')
plt.show()

```

```python
# 2. Most Common Genres
# Split the 'listed_in' column and count genres
dataframe['genres'] = dataframe['listed_in'].apply(lambda x: x.split(','))
all_genres = sum(dataframe['genres'], [])
genre_counts = pd.Series(all_genres).value_counts().head(10)
# Plot the most common genres
plt.figure(figsize=(10, 6))
sns.barplot(x=genre_counts.values, y=genre_counts.index,
palette='Set3')
plt.title('Most Common Genres on Netflix')
plt.xlabel('Count')
plt.ylabel('Genre')
plt.show()
```

```python
# 3. Content Added Over Time
# Extract year and month from 'date_added'
dataframe['year_added'] = dataframe['date_added'].dt.year
dataframe['month_added'] = dataframe['date_added'].dt.month

# Plot content added over the years
plt.figure(figsize=(12, 6))
sns.countplot(x='year_added', data=dataframe, palette='coolwarm')
plt.title('Content Added Over Time')
plt.xlabel('Year')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# 4. Top 10 Directors with the Most Titles
# Count titles by director
top_directors = dataframe['director'].value_counts().head(10)
# Plot top directors
plt.figure(figsize=(10, 6))
sns.barplot(x=top_directors.values, y=top_directors.index,
palette='Blues_d')
plt.title('Top 10 Directors with the Most Titles')
plt.xlabel('Number of Titles')
plt.ylabel('Director')
plt.show()
```

```python
# 5. Word Cloud of Movie Titles
# Generate word cloud
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import pandas as pd
dataframe = pd.read_csv("netflix_titles.csv")  # or however you're loading your data


movie_titles = dataframe[dataframe['type'] == 'Movie']['title']
wordcloud = WordCloud(width=800, height=400,
background_color='black').generate(' '.join(movie_titles))

# Plot word cloud
plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
